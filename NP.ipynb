```python
import re
```

```python
class Uzel:
    
    """
    класс узел в дереве. 
    Хранит 4 параметра: 
    1) id, который одновременно является указателем порядка в предложении
    2) parent -- id узла родителя
    3) 
    4) 
    """
    
    def __init__(self, id_number, word, parent, pos):
        self.id_number = id_number
        self.word = word
        self.parent = parent
        self.pos = pos
```

```python
def parse_file (path_to_file):
    with open (path_to_file, 'r') as conllfile:
        sentence_id = '0'
        for line in conllfile:
            if line.startswith('# sent_id'):
                sentence_id = re.search ('sent_id = (\d+)', line).group(1)
            elif re.match ('\d', line):
                add_line_with_node_to_all_sent (line, sentence_id)
```

```python
def add_line_with_node_to_all_sent (some_line, sentence_id):
        new_node = parse_node (some_line)
        if sentence_id in all_sentences:
            all_sentences[sentence_id].append(new_node)
        else:
            all_sentences[sentence_id] = [new_node]
```

```python
def parse_node (some_line):
    node_features = some_line.split('\t')
    id_number = node_features[0]
    word = node_features[1]
    pos = node_features[3]
    parent = node_features[6]
    new_node = Uzel (id_number,word, parent, pos)
    return (new_node)
```

```python
noun_phrases = [] # сюда сложим именные группы
```

```python
def get_noun_phrases (sentences_dict):
    for sentence_index in sentences_dict:
        sentence_nodes = sentences_dict [sentence_index]
        for node in sentence_nodes:
            if node.pos == 'NOUN':
                this_node_and_descendents = []
                this_node_and_descendents.append (node)
                noun_phrase = get_descendents (node, sentence_nodes, this_node_and_descendents)
                noun_phrases.append (noun_phrase)
            
```

```python
def get_descendents (this_node, sentence_nodes, this_node_and_descendents):
    for some_node in sentence_nodes:
        if some_node.parent == this_node.id_number:
            this_node_and_descendents.append  (some_node)
            get_descendents (some_node, sentence_nodes, this_node_and_descendents)
    return this_node_and_descendents
```

```python
def print_ordered_noun_phrases (some_noun_phrases_list):
    for phrase in some_noun_phrases_list:
        # отсортируем по id_number, который у нас кодирует исходный порядок слов:
        sorted_phrase = sorted(phrase, key = lambda item: item.id_number)
        sorted_phrase_text = []

        for node in sorted_phrase:
        #sorted_phrase_text.append (node.word)
            sorted_phrase_text.append (node.id_number)

        print (' '.join (sorted_phrase_text))
```
